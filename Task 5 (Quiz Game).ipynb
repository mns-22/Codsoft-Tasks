```python
import random

# Define a list of quiz questions and answers.
quiz_questions = [
    {
        "question": "What is the capital of France?",
        "choices": ["A) Paris", "B) London", "C) Berlin", "D) Madrid"],
        "correct_answer": "A"
    },
    {
        "question": "Which planet is known as the Red Planet?",
        "choices": ["A) Venus", "B) Mars", "C) Jupiter", "D) Saturn"],
        "correct_answer": "B"
    },
    {
        "question": "What is the largest mammal in the world?",
        "choices": ["A) Elephant", "B) Blue Whale", "C) Giraffe", "D) Lion"],
        "correct_answer": "B"
    }
]

def display_welcome_message():
    print("Welcome to the Quiz Game!")
    print("You will be asked multiple-choice questions. Choose the correct option.")
    print("Let's begin!\n")

def ask_question(question_dict):
    print(question_dict["question"])
    for choice in question_dict["choices"]:
        print(choice)
    user_answer = input("Your answer (A/B/C/D): ").strip().upper()
    return user_answer

def evaluate_answer(question, user_answer):
    correct_answer = question["correct_answer"]
    if user_answer == correct_answer:
        print("Correct!\n")
        return 1
    else:
        print(f"Sorry, the correct answer is {correct_answer}.\n")
        return 0

def main():
    display_welcome_message()
    play_again = True
    total_score = 0

    while play_again:
        random.shuffle(quiz_questions)
        current_score = 0

        for question in quiz_questions:
            user_answer = ask_question(question)
            current_score += evaluate_answer(question, user_answer)

        total_score += current_score
        print(f"Your score for this round: {current_score}/{len(quiz_questions)}")
        
        play_again_input = input("Do you want to play again? (yes/no): ").strip().lower()
        play_again = (play_again_input == "yes")

    print(f"Your final score: {total_score}/{len(quiz_questions) * len(quiz_questions)}")
    print("Thank you for playing!")

if __name__ == "__main__":
    main()
```
